Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
def get_owner_name(documents):
    """
    Возращает имя владельца по номеру документа
    """
    doc_owner = ''
    doc_num = input('Введите номер документа ')
    for document in documents:
        if document['number'] == doc_num:
            doc_owner = document['name']
            break
    if doc_owner != '':
        print(f"Владелец: {doc_owner}")
    else:
        print('Документ не найден')


def get_shelf_num(directories):
    """
    Возвращает номер полки по номеру документа на ней
    """
    doc_num = input('Введите номер документа ')
    doc = 0
    for key, directory in directories.items():
        if doc_num in directory:
            doc = key
    if doc != 0:
        print(f"Полка: {doc}")
    else:
        print('Документ не найден')

def get_doc_list(documents):
    """
    Возвращает список документов
    """
    for document in documents:
        print(f"{document['type']} \"{document['number']}\" \"{document['name']}\"")
        
def add_shelf(directories):
    """
    Добавляет новую полку
    """
    shelf_num = input('Введите номер полки ')
    if str.isnumeric(shelf_num):
        if shelf_num in directories:
            print('Такая полка уже есть')
        else:
            directories.setdefault(shelf_num, [])
            print('Полка добавлена')
    else:
        print('Некорректный номер полки')
        
def delete_doc(documents, directories):
    """
    Удаляет документ из каталога и из полки
    """
    doc_num = input('Введите номер документа ')

    new_documents = []
    c = 0
    for n, document in enumerate(documents):
        if document['number'] == doc_num:
            c += 1
        else:
            new_documents.append(document)
    if c == 0:
        print('Документ не найден')  
    else:
        documents = new_documents
        print('Документ удален')        

    i = 0
    for key, directory in directories.items():
        if doc_num in directory:
            i += 1
            directory.remove(doc_num)
    if i != 0:
        print('Документ удален с полки')
    else:
        print('Полка не найдена')
        
def move_doc(directories):
    """
    Перемещает документ на новую полку
    """
    shelf_list = []
    for shelf in directories.keys():
        shelf_list.append(shelf)

    doc_num = input('Введите номер документа ')
    shelf_num = input('Введите номер полки ')
    d = 0
    for key, directory in directories.items():
        if doc_num in directory:
            d += 1
            if shelf_num in shelf_list:
                directory.remove(doc_num)
            else:
                print('Полка не найдена')
                break

    if d == 0:
        print('Документ не найден')
    else:
        for key1, directory1 in directories.items(): 
            if shelf_num == key1:
                directory1.append(doc_num)
                print('Документ перемещен')
                break
                
def add_doc(documents, directories):
    """
    Добавляет документ в каталок и на указанную полку
    """
    new_dict = {}
    new_dict['type'] = input('Введите тип документа ')
    new_dict['number'] = input('Введите номер документа ')
    new_dict['name'] = input('Введите владельца документа ')
    documents.append(new_dict)

    shelf_num = input('Введите номер полки ')

    shelf_list = []
    for shelf in directories.keys():
        shelf_list.append(shelf)

    if shelf_num in shelf_list:
        for key, directory in directories.items():
            if key == shelf_num:
                directory.append(new_dict['number'])
                break
    else:
        directories[shelf_num] = new_dict['number']
```

```python
def main():
    while True:
        user_input = input('Введите команду ')
        if user_input == 'p':
            get_owner_name(documents)
        elif user_input == 's':
            get_shelf_num(directories)
        elif user_input == 'l':
            get_doc_list(documents)
        elif user_input == 'as':
            add_shelf(directories)
        elif user_input == 'd':
            delete_doc(documents, directories)
        elif user_input == 'm':
            move_doc(directories)
        elif user_input == 'a':
            add_doc(documents, directories)
        elif user_input == 'q':
            print('До новых встреч!')
            break
            
main()
```
