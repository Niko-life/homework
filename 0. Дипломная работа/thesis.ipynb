# Дипломная работа

```python
import pandas as pd
from IPython.display import Image
import statistics as sts
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

1. Загрузить файл hr.csv

```python
Image('описание.png')
```

```python
data = pd.read_csv('HR.csv')
data.head()
```

2. Рассчитать основные статистики для переменных (среднее, медиана, мода, мин/макс, сред. отклонение)

```python
data.describe()
```

```python
for i in data.columns[:-2]:
    print('Переменная:', i)
    print('Медиана: ', sts.median(data[i]))
    print('Мат. ожидание: ', sts.mean(data[i]))
    print('Станд. отклонение: ', sts.pstdev(data[i]))
    print('Дисперсия: ', sts.pvariance(data[i]))
    print('Минимум: ', min(data[i]))
    print('Максимум: ', max(data[i]))
    try:
        print('Мода: ', sts.mode(data[i]))
    except:
        print('Мода: невозможно найти моду')
    print('-------------------------------------')
```

3. Рассчитать и визуализировать корреляционную матрицу для переменных

```python
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
fig = plt.gcf()
fig.set_size_inches(8, 8)
plt.show()
```

**Вывод:** \
Слабая корреляция между большинством параметров. Наиболее выделяются:

* number_project и average_montly_hours - слабая положительная корреляция
* satisfaction_level и left - слабая отрицательная корреляция
* last_evaluation и average_montly_hours - слабая положительная корреляция

4. Визуализировать несколько взаимосвязей между переменными

```python
number_project = data['number_project']
average_montly_hours = data['average_montly_hours']
satisfaction_level = data['satisfaction_level']
left = data['left']
last_evaluation = data['last_evaluation']
```

```python
# Взаимосвязь number_project и average_montly_hours

fig, ax = plt.subplots()
ax.scatter(average_montly_hours, number_project, alpha=0.3, edgecolors='none')
ax.grid(True)
plt.title('number_project и average_montly_hours')
plt.xlabel('average_montly_hours')
plt.ylabel('number_project')
plt.show()

# Взаимосвязь satisfaction_level и left

fig, ax = plt.subplots()
ax.scatter(satisfaction_level, left, alpha=0.3, edgecolors='none')
ax.grid(True)
plt.title('satisfaction_level и left')
plt.xlabel('satisfaction_level')
plt.ylabel('left')
plt.show()

# Взаимосвязь last_evaluation и average_montly_hours

fig, ax = plt.subplots()
ax.scatter(last_evaluation, average_montly_hours, alpha=0.3, edgecolors='none')
ax.grid(True)
plt.title('last_evaluation и average_montly_hours')
plt.xlabel('last_evaluation')
plt.ylabel('average_montly_hours')
plt.show()
```

**Вывод:** 
* number_project и left являются дискретными параметрами
* корреляцию между last_evaluation и average_montly_hours можно рассматривать как случайную

5. С использованием t-test статистики проверить взаимосвязи между несколькими переменными

```python
# Взаимосвязь number_project и average_montly_hours

res1 = stats.ttest_ind(number_project, average_montly_hours)
print('t-value для number_project и average_montly_hours:', res1.statistic)

# Взаимосвязь satisfaction_level и left

res2 = stats.ttest_ind(satisfaction_level, left)
print('t-value для satisfaction_level и left:', res2.statistic)

# Взаимосвязь last_evaluation и average_montly_hours

res3 = stats.ttest_ind(last_evaluation, average_montly_hours)
print('t-value для last_evaluation и average_montly_hours:', res3.statistic)
```

**Вывод:** выборки со средним уровнем корреляции Пирсона могут значительно отличаться друг от друга

10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

```python
# Гипотеза: чем выше average_montly_hours, тем больше вероятность покинуть компанию

amh_stayed = data[data['left']==0]['average_montly_hours']
amh_left = data[data['left']==1]['average_montly_hours']

print(stats.ttest_ind(amh_stayed, amh_left))

sns.distplot(amh_stayed, 50)
sns.distplot(amh_left, 50)
plt.title('Плотность вероятности распределения average_montly_hours')
plt.legend(['остались в компании', 'покинули компанию'])
plt.show()
```

**Вывод:** \
p-value гораздо ниже 0.05 - отвергаем $H_0$, выборки отличаются друг от друга. \
Параметр может иметь влияние на решение сотрудников об уходе. \
Покинувшие компанию сотрудники работали либо меньше, либо больше относитьельно среднего значения.

```python
# Гипотеза: чем ниже satisfaction_level, тем больше вероятность покинуть компанию

sl_stayed = data[data['left']==0]['satisfaction_level']
sl_left = data[data['left']==1]['satisfaction_level']

print(stats.ttest_ind(sl_stayed, sl_left))

sns.distplot(sl_stayed, 50)
sns.distplot(sl_left, 50)
plt.title('Плотность вероятности распределения satisfaction_level')
plt.legend(['остались в компании', 'покинули компанию'])
plt.show()
```

**Вывод:** \
p-value гораздо ниже 0.05 - отвергаем $H_0$, выборки отличаются друг от друга. \
Параметр может иметь влияние на решение сотрудников об уходе. \
Покинувшие компанию сотрудники имели низкий уровень удовлетворения работой относитьельно среднего значения.

```python
# Гипотеза: чем ниже salary, тем больше вероятность покинуть компанию

df = data
df['salary_index'] = df.apply(lambda x: 1 if x['salary']=='low' else (2 if x['salary']=='medium' else 3), axis=1)

s_stayed = df[df['left']==0]['salary_index']
s_left = df[df['left']==1]['salary_index']

print(stats.ttest_ind(s_stayed, s_left))

sns.distplot(s_stayed, 50)
sns.distplot(s_left, 50)
plt.title('Плотность вероятности распределения salary')
plt.legend(['остались в компании', 'покинули компанию'])
plt.show()
```

**Вывод:** \
p-value гораздо ниже 0.05 - отвергаем $H_0$, выборки отличаются друг от друга. \
Параметр может иметь влияние на решение сотрудников об уходе. \
Покинувшие компанию сотрудники имели уровень заработка ниже относитьельно среднего значения.
